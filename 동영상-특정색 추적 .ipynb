```python
import cv2
import numpy as np

color = [255,0,0]
pixel = np.uint8([[color]])

hsv = cv2.cvtColor(pixel,cv2.COLOR_BGR2HSV)
hsv = hsv[0][0]
lower_blue = (150-50,0,0)
upper_blue = (150+50,255,255)

cap= cv2.VideoCapture('blueball2.mp4')

while cap.isOpened():
    ret,img_color = cap.read()
    img_color = cv2.resize(img_color,(640,480),interpolation=cv2.INTER_AREA)
    
    if not ret:
        print('x')
        break
    
    img_hsv = cv2.cvtColor(img_color,cv2.COLOR_BGR2HSV)
    img_mask = cv2.inRange(img_hsv,lower_blue,upper_blue)

    img_result = cv2.bitwise_and(img_color,img_color,mask = img_mask)
    
    _,img_label,stats,centroids = cv2.connectedComponentsWithStats(img_mask)
    
    for idx,centroid in enumerate(centroids):
        if stats[idx][0] == 0 and stats[idx][1]==0:
            continue
        if np.any(np.isnan(centroid)):
            continue
        x,y,width,height,area = stats[idx]
        centerX,centerY = int(centroid[0]),int(centroid[1])
        print(centerX,centerY)
        
        if area>70:
            cv2.circle(img_color,(centerX,centerY),5,(0,255,0),5)
            cv2.rectangle(img_color,(x,y),(x+width,y+height),(0,255,0))
            cv2.circle(img_result,(centerX,centerY),5,(0,0,155),5)
            cv2.rectangle(img_result,(x,y),(x+width,y+height),(0,0,255))
    
    cv2.imshow('img_color',img_color)
    cv2.imshow('img_result',img_result)
    
    if cv2.waitKey(100) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
